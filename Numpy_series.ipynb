Operations with the numpy

*   Numpy array supports vectorisation which means they can perform operations on entire groups of numbers at once, which is called vectorized operations.



```python
l = [1,2,3]
print("Normal list : " , l)
arr = np.array([1,2,3])
print("Numpy array : ", arr)
```

```python
# 2-D array
two_d_array = np.array([[1, 2, 3], [4, 5, 6]])
print("2-D array")
print(two_d_array)
```

Indexing

```python
indexing_in_numpy = two_d_array[0,1] # where [0,1] determines the indexing value
# like 0 refers to the first array and 1 refers to the arr[0][1]th element in the nested array
print(indexing_in_numpy)
```

Accessing the rows and columns

```python
row = two_d_array[0 , :] #[starting index,  (nested array starting : nested array ending) of row]
print(row)
column = two_d_array[: , 0] ##[ (nested array starting : nested array ending )of column , ending index ]
print(column)
```

Copying a np.array

```python
copy_array = two_d_array.copy()
print("copy of the 2-d string")
print(copy_array)
```

**Copy v/s View**
<br>
Copy() - Changes made to the original array do not affect the copied array, and vice versa. They are entirely independent.

View() - A view in NumPy is a different way of looking at the same data without creating a new array.

```python
view_array = two_d_array.view()
print("View of the array")
print(view_array)
```

```python
transpose_of_array = two_d_array.T
# transpose of the array
print("Transpose of the array : ")
print(transpose_of_array)
```

```python
print("Reshape the array")
reshaped_Array = two_d_array.reshape((6,1))
print(reshaped_Array)
```

# **Mathematical Functions**

```python
print("Sum of the array")
sum = np.sum(two_d_array)
sum_by_col = np.sum(two_d_array , axis=0)  # Sum by columns
sum_by_row = np.sum(two_d_array , axis=1)  # Sum by rows
print(sum , sum_by_row , sum_by_col)
```

```python
print("Multiplying the np array 2 ")
multiplied = np.multiply(two_d_array, 2)
print(multiplied)
```

```python
print("Power of the array elements to be 3")
expo = np.power(two_d_array , 3)
print(expo)
```

# Statistical Functions

```python
print("Mean of the 2-D array ")
mean = np.mean(two_d_array)
print(mean)
```

```python
print("Standard deviation of 2-D array")
std_dev = np.std(two_d_array)
print(std_dev)
```

# Boardcasting
#### The ability to perform arithmetic operations on arrays of different shapes is called boardcasting . It is possible if the shapes are compatiable .

```python
bs = np.arange(5)
print("array is" , bs)
```

```python
bs + 2 # after adding 2 to the array , the result is given below
```

```python
bs1 = np.arange(10).reshape((2,5))
bs1
```

```python
bs + bs1 # here , bs has 1 row and bs1 has two rows . to make the arithmetic operation , the row of bs is duplicated and added as second row and thus it is added with bs1 .
```

```python
bs * 2 # after multiplying 2 to it , the result is given below
```

```python
bs2 = np.arange(5).reshape((1,5))
bs3 = np.arange(5).reshape((5,1))
print("bs1 is" , bs2)
print("bs2 is" , bs3)
```

```python
bs2 + bs3  #here , both the array try to make compatiable with each other to make the arithmetic operation
```
